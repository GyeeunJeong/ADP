# SNA
- 노드, 엣지(연결선)으로 이루어짐
- 노드의 색, 모양 그리고 엣지의 두께 등으로 속성 표현
- 연결선은 방향성(A>B)과 비방향성(A, B가 만난 횟수 등)이 있음

### 용어 정의
- Degree: 노드가 갖는 연결선의 수. 방향성 있다면 입선(indegree), 출선(outdegree)으로 나눌 수 있음
  - 입선: 들어오는 선, 출선: 나가는 선
- 밀도(Density): 모든 노드의 입선 수(출선 수)의 합 / 가능한 연결선의 모든 경우의 수
  - $\frac{\Sigma^n_{i=1}g_i}{n(n-1)}$
- 도달 가능도(Reachability): 노드 쌍에 대하여 한 노드에서 다른 노드로 도달 가능한지 여부를 0 또는 1로 나타냄 (from~to까지의 길)
  - 간접적으로라도 가능하면 1
- 컴포넌트(Component): 내부적으로 상호 도달 가능한 노드들의 그룹. DBSCAN군집과 유사.
  - 컴포넌트 개수 정해지면 사이즈는 자동으로 결정
  - 컴포넌트가 작을수록 좁은 세상
- 중심성(Centrality)
  - 네트워크에서 가장 중요한 노드는?
  - 근접 중심성: Closeness Centrality. 해당 노드와 다른 모든 노드와의 평균 최단거리를 구한 다음 역수 취함. 기하학적 중심.
    - 평균 최단거리가 작을수록 네트워크의 중심에 위치.
  - 중개 중심성: Betweeness Centrality. 네트워크에서 허브역할을 얼마나 잘하는지, 정보의 통로라 전달 역할을 얼마나 잘하는지.
    - 전염병 확산, 소문, 마케팅 확장 등의 중요한 변수.
    - $C(V)=\sum_i\sum_j{\frac{shortest_{ivj}}{shortest_{ij}}}$  
    - $shortest_{ivj}$: i에서 j로 가는데 v를 거쳐 가는 경로
  - 아이겐벡터 중심성: Eigenvector Centrality. 연결된 노드에 따라 가중치를 부여.
    - 중요한 노드에 연결된 노드는 더 많은 중요성을 부여.
    - 인접행렬 $A(=a_{ij})의 n개의 노드에 대해서 $λc=Ac$를 푸는 고유벡터
    - $a_{ij}$: 엣지의 가중치, $c_j$: 노드의 중요도, $λ$: 비례상수

```python
import csv
from operator import itemgetter
import community
```

```python
with open('ADP/data/quakers_nodelist.csv', 'r') as nodecsv: # 노드 데이터
  nodereader = csv.reader(nodecsv)

  nodes = [n for n in nodereader][1:]                        # 헤더정보를 빼고 가져온다
node_names = [n[0] for n in nodes]                           # 노드 이름 가져옴

with open('ADP/data/quakers_edgelist.csv', 'r') as edgecsv:  # 노드 간 연결정보
  edgereader = csv.reader(edgecsv)
  edges = [tuple(e) for e in edgereader][1:]                 # 튜플정보로 변형한다.
```

```python
edgereader, edges[:10] # George Keith > Robert Barclay. 강도 정보는 없어 모두 같은 것으로 취급
```

```python
nodes[:5]
```

```python
node_names[:5]
```

```python
len(node_names), len(edges)
```

```python
G = nx.Graph()                # 그래프 객체 초기화
G.add_nodes_from(node_names)  # G객체에 노드 추가
G.add_edges_from(edges)       # G객체에 엣지 추가
nx.info(G)
```

```python
# 노드 속성정보를 사전 형태로 만듬
hist_sig_dict = {}
gender_dict = {}
birth_dict = {}
death_dict = {}
id_dict = {}
```

```python
for node in nodes:
  hist_sig_dict[node[0]] = node[1]
  gender_dict[node[0]] = node[2]
  birth_dict[node[0]] = node[3]
  death_dict[node[0]] = node[4]
  id_dict[node[0]] = node[5]
```

```python
nx.set_node_attributes(G, hist_sig_dict, 'historical_significance')
nx.set_node_attributes(G, gender_dict, 'gender')
nx.set_node_attributes(G, birth_dict, 'birth_year')
nx.set_node_attributes(G, death_dict, 'death_year')
nx.set_node_attributes(G, id_dict, 'sdfb_id')
```

```python
G.nodes()
```

```python
for n in G.nodes():
  print(n, G.nodes[n]['birth_year'])
```

```python
# 네트워크의 밀도
density = nx.density(G)
print('Network density:', density)
```

```python
# from > to의 최단거리
fell_whitehead_path = nx.shortest_path(G, source='Margaret Fell', target='George Whitehead')
print('Shortest path between Fell and Whitehead:', fell_whitehead_path)
```

```python
# 위에서 구한 최단거리의 길이
print('Length of that path:', len(fell_whitehead_path)-1)
```

```python
# 네트워크의 컴포넌트는 1개인가? 즉 모두 연결이 되어 있는가?
print(nx.is_connected(G))

# 컴포넌트의 리스트
# 가장 큰 컴포넌트
components = nx.connected_components(G)
largest_component = max(components, key=len)

# 가장 큰 컴포넌트의 서브그래프를 만든다
# 서브그래프의 반경(크기)를 출력
subgraph = G.subgraph(largest_component)
diameter = nx.diameter(subgraph)
print('Network diameter of largest component:', diameter) # 8까지함
```

```python
# 이행성의 비율. A>B, B>C가 있을 때 A>C의 비율
triadic_closure = nx.transitivity(G)
print('Triadic closure:', triadic_closure)
```

```python
# 각각의 노드의 degree를 계산하여 노드의 속성정보로 세팅
degree_dict = dict(G.degree(G.nodes()))
nx.set_node_attributes(G, degree_dict, 'degree')
```

```python
# William Penn의 노드정보
print(G.nodes['William Penn'])
```

```python
# 모든 노드의 degree를 내림차순으로 나타냄
sorted_degree = sorted(degree_dict.items(), key=itemgetter(1), reverse=True) #itemgetter(0)은 degree_dict의 key, (1)은 degree_dict의 value
```

```python
sorted_degree[:5]
```

```python
# 상위20순위의 degree
sorted_degree[:20]
```

```python
from networkx.algorithms.bipartite.centrality import betweenness_centrality
betweenness_dict = nx.betweenness_centrality(G)  # 중개중심성
eigenvector_dict = nx.eigenvector_centrality(G)  # 아이겐벡터중심성 # 두 중심성이 서로 다른 패턴

# 각각의 노드의 속성에 위에서 계산한 중심성을 세팅
nx.set_node_attributes(G, betweenness_dict, 'betweenness')
nx.set_node_attributes(G, eigenvector_dict, 'eigenvector')
```

```python
# 상위 20개의 중개중심성 출력
sorted_betweenness = sorted(betweenness_dict.items(), key=itemgetter(1), reverse=True)

print('Top 20 nodes by betweenness centrality:')
for b in sorted_betweenness[:20]:
  print(b)
```

```python
# 상위 20개의 중개중심성을 추출
top_betweenness = sorted_betweenness[:20]

# 중개중심성 상위 20개를 가져와서 노드의 이름, degree와 함께 출력
for tb in top_betweenness:
  degree = degree_dict[tb[0]]
  print('Name:', tb[0], '| Betweenness Centrality:', tb[1], '| Degree:', degree) # 이 그래프는 중개중심성과 degree의 상관관계가 강한 편이다
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

G_fb = G

# 노드의 색은 degree의 정도이고 노드의 반지름은 중개 중심성 지수이다.
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0*G_fb.degree(v) for v in G_fb]
node_size = [v*10000 for v in betCent.values()]

plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=True, node_color=node_color, node_size=node_size)
plt.axis('off')
```

## 실습2

```python
import pandas as pd
df = pd.read_csv('./data/smartphone.csv', encoding='utf-8')
galexy_posts = df.get('Title') + " " + df.get('Description')
galexy_post_date = df.get('Post Date')
```

```python
from eunjeon import Mecab
tagger = Mecab()

galexy_stop_words = "은 이 것 등 더 를 좀 즉 인 옹 때 만 원 이때 개 일 기 시 럭 갤 성 삼 스 폰 트 드 기 이 리 폴 사 전 마 자 플 블 가 중 북 수 팩 년 월 저 탭"
galexy_stop_words = galexy_stop_words.split(' ')
galexy_stop_words[0:10]
```

```python
galexy_nouns = []
for post in galexy_posts:
    for noun in tagger.nouns(post):
        if noun not in galexy_stop_words:
            galexy_nouns.append(noun)
            
galexy_nouns[0:10]
```

```python
from collections import Counter
num_top_nouns = 20
galexy_nouns_counter = Counter(galexy_nouns)
galexy_top_nouns = dict(galexy_nouns_counter.most_common(num_top_nouns))
```

```python
galexy_sentences = []
for post in galexy_posts:
    galexy_sentences.extend(re.split('; |\.|\?|\!', post))
galexy_sentences[0:10]
```

```python
galexy_sentences_nouns = []
for sentence in galexy_sentences:
    sentence_nouns = tagger.nouns(sentence)
    galexy_sentences_nouns.append(sentence_nouns)
galexy_sentences_nouns[0:10]
```

```python
galexy_word2id = {w: i for i, w in enumerate(galexy_top_nouns.keys())}
galexy_word2id
```

```python
galexy_id2word = {i: w for i, w in enumerate(galexy_top_nouns.keys())}
galexy_id2word
```

```python
import numpy as np
galexy_adjacent_matrix = np.zeros((num_top_nouns, num_top_nouns), int)
for sentence in galexy_sentences_nouns:
    for wi, i in galexy_word2id.items():
        if wi in sentence:
            for wj, j in galexy_word2id.items():
                if i != j and wj in sentence:
                    galexy_adjacent_matrix[i][j] += 1
galexy_adjacent_matrix
```

```python
galexy_network = nx.from_numpy_matrix(galexy_adjacent_matrix)
list(galexy_network.adjacency())
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
from matplotlib import rc

font_path="./font/NanumBarunGothic.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

fig = plt.figure()
fig.set_size_inches(20, 20)
ax = fig.add_subplot(1, 1, 1)
ax.axis("off")
option = {
    'node_color' : 'lightblue',
    'node_size' : 2000,
    'size' : 2
}
nx.draw(galexy_network, labels=galexy_id2word, font_family=font_name, ax=ax, **option)
```

```python
fig = plt.figure()
fig.set_size_inches(20, 20)
option = {
    'node_color' : 'lightblue',
    'node_size' : 500,
    'size' : 100
}

plt.subplot(221)
plt.title('Random Layout', fontsize=20)
nx.draw_random(galexy_network, labels=galexy_id2word, font_family=font_name, **option)
plt.subplot(222)
plt.title('Circular Layout', fontsize=20)
nx.draw_circular(galexy_network, labels=galexy_id2word, font_family=font_name, **option)
plt.subplot(223)
plt.title('Spectral Layout',fontsize=20)
nx.draw_spectral(galexy_network, labels=galexy_id2word, font_family=font_name, **option)
plt.subplot(224)
plt.title('Spring Layout',fontsize=20)
nx.draw_spring(galexy_network, labels=galexy_id2word, font_family=font_name, **option)
```
